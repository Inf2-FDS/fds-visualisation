## Loading in packages

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

## Set default figure parameters

Setting the figure size to be 4 inches wide and 3 inches high with a resolution of 200 dots per inch will make raster images that are 800 pixels wide and 600 pixels high.

```python
mpl.rcParams['figure.dpi'] = 200
mpl.rcParams['figure.figsize'] = (4, 3)  # 4 inches wide and 3 inches high
```

## Load the data

L. A. Wauters and A. A. Dhondt, Variation in length and body weight of the red squirrel (Sciurus vulgaris) in two different habitats, J. Zool. 217 (1989), 93â€“106, DOI: 10.1111/j.1469-7998.1989.tb02477.x.

```python
dat = pd.read_csv('squirrel-clean.csv')
```

## Font sizes

This plot uses 10pt font size, but because the figure axes is sized to be 6in wide,  and the text overhangs the side of the figure, when shrunk to 6in wide, the resulting font size will be smaller than 10pt.

```python
mpl.rcParams['font.size'] = 10
plt.figure(figsize=(6, 3)) # This figure is 6 inches wide and 3 inches tall
ax = sns.boxplot(x='Year', y='Weight (g)', hue='Sex', data=dat)
plt.title('Distribution of weights of male and female squirrels in 1985 and 1986 measured by Wauters and Dhondt (1989)')
plt.ylim([275, 450])
sns.move_legend(ax, 'upper right')
plt.tight_layout()
plt.savefig('python-6x3-title.png')
plt.savefig('python-6x3-title.pdf')
```

If we make the figure size bigger, it makes more space to fit in the text... but then the figure will be shrunk to a width of 6 inches in the PDF output, so the font will reduce in size. There the shrinkage is by a factor of 3.

```python
mpl.rcParams['font.size'] = 10
plt.figure(figsize=(18, 9))
ax = sns.boxplot(x='Year', y='Weight (g)', hue='Sex', data=dat)
plt.title('Distribution of weights of male and female squirrels in 1985 and 1986 measured by Wauters and Dhondt (1989)')
plt.savefig('python-18x9.png')
plt.savefig('python-18x9.pdf')
```

Setting the figure width back to 6in, and putting a line break in the title means that the true font size will be 10pt when the figure is scaled to be 6in wide.

```python
mpl.rcParams['font.size'] = 10
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x='Year', y='Weight (g)', hue='Sex', data=dat)
# Note the line break entered using '\n'
plt.title('Distribution of weights of male and female squirrels in\n1985 and 1986 measured by Wauters and Dhondt (1989)')
plt.ylim([275, 450])
sns.move_legend(ax, 'upper right')
plt.tight_layout()
plt.savefig('python-6x3.png')
plt.savefig('python-6x3.pdf')
```

If we make the figure size bigger, it makes more space to fit in the text... but then the figure will be shrunk to a width of 6 inches in the PDF output, so the font will reduce in size. There the shrinkage is by a factor of 3.

```python
mpl.rcParams['font.size'] = 10
plt.figure(figsize=(18, 9))
ax = sns.boxplot(x='Year', y='Weight (g)', hue='Sex', data=dat)
plt.title('Distribution of weights of male and female squirrels in 1985 and 1986 measured by Wauters and Dhondt (1989)')
plt.savefig('python-18x9.png')
plt.savefig('python-18x9.pdf')
```

## Setting the font size to 8 point

This is the smallest size we recommend.

```python
mpl.rcParams['font.size'] = 8
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x='Year', y='Weight (g)', hue='Sex', data=dat)
# Note the line break entered using '\n'
plt.title('Distribution of weights of male and female squirrels in\n1985 and 1986 measured by Wauters and Dhondt (1989)')
plt.ylim([275, 450])
sns.move_legend(ax, 'upper right')
plt.tight_layout()
plt.savefig('python-6x3-8pt.png')
plt.savefig('python-6x3-8pt.pdf')
```
